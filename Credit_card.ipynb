```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix
from sklearn.cluster import KMeans
import tensorflow as tf
import seaborn as sns
from google.colab import files

# Load the data from 'output (2).csv'. This file should already have correct headers.
df = pd.read_csv("output (2).csv")

df = df.drop_duplicates()

# Convert 'TransactionDate' to datetime objects and extract relevant features
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['TransactionMonth'] = df['TransactionDate'].dt.month
df['TransactionDay'] = df['TransactionDate'].dt.day
df['TransactionHour'] = df['TransactionDate'].dt.hour
df = df.drop('TransactionDate', axis=1) # Drop original date column

# Define features (X) and target (y) based on the customer dataset
# 'IsFraud' is the target variable for this dataset
X = df.drop("IsFraud", axis=1)
y = df["IsFraud"]

# Identify categorical features for the customer dataset
categorical_features = ['TransactionType', 'Location']

# Handle NaNs and encode categorical features
for col in X.columns:
    if X[col].isnull().any():
        if pd.api.types.is_numeric_dtype(X[col]):
            # Fill numeric NaNs with median, if median is NaN, fill with 0
            median_val = X[col].median()
            X[col] = X[col].fillna(median_val if not pd.isna(median_val) else 0)
        else:
            # For categorical columns, fill NaNs with a placeholder before encoding
            X[col] = X[col].fillna('__MISSING__')

for col in categorical_features:
    if col in X.columns:
        le = LabelEncoder()
        # Ensure the column is of object type if it contains string placeholders
        X[col] = X[col].astype(str)
        X[col] = le.fit_transform(X[col])

# Scale numerical features
# First, identify numerical columns that are not already label encoded
numerical_cols = [col for col in X.columns if col not in categorical_features and pd.api.types.is_numeric_dtype(X[col])]

# Apply StandardScaler only to numerical columns
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

model.save("tf_model.keras")

y_prob = model.predict(X_test).ravel()
y_pred = (y_prob > 0.5).astype(int)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.show()

# KMeans clustering is usually applied to features, not the original df mixed with target
# If you intend to cluster the processed features:
kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto')
df_clustered = X.copy() # Create a copy of processed features for clustering
df_clustered["cluster"] = kmeans.fit_predict(X) # Cluster on processed features

# Visualizations using the actual customer data columns
plt.figure(figsize=(8, 6))
sns.countplot(x=y) # Use 'y' for the actual distribution of IsFraud
plt.title('Distribution of IsFraud (Target Variable)')
plt.xlabel('IsFraud (0: Not Fraud, 1: Fraud)')
plt.ylabel('Count')
plt.show()

# Example scatter plot: Amount vs Transaction Hour by Fraud Status
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df['Amount'], y=df['TransactionHour'], hue=y, data=df.sample(n=min(5000, len(df)), random_state=42))
plt.title('Scatter Plot of Amount vs Transaction Hour by Fraud Status')
plt.xlabel('Amount')
plt.ylabel('Transaction Hour')
plt.show()

# Example histogram: Distribution of Transaction Amount
plt.figure(figsize=(10, 6))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.yscale('log') # Log scale might be useful for skewed financial data
plt.show()
```
